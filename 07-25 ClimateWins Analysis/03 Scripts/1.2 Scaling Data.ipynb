## 3. Scaling

```python
#create scaler object using StandardScaler from sklearn.preprocessing
#StandardScaler assumes data is normally distributed and scales with a distribution
#around 0 and standard deviation of 1. Scaling happens independently with each variable.

scaler = StandardScaler()
```

```python
#create new df with scaled data

df_scaled = pd.DataFrame(scaler.fit_transform(df_subset), columns=df_subset.columns)
df_scaled.head()
```

```python
# Export scaled data
df_scaled.to_csv(os.path.join(path, 'Prepared Data', 'weather_prediction_scaled.csv'))
```

```python
# Add DATE and MONTH back to the scaled dataset
df_scaled['DATE'] = df['DATE'].values
df_scaled['MONTH'] = df['MONTH'].values

# move DATE and MONTH to the front
cols = ['DATE', 'MONTH'] + [col for col in df_scaled.columns if col not in ['DATE', 'MONTH']]
df_scaled = df_scaled[cols]
```

```python
# Export new dataframe 
df_scaled.to_csv(os.path.join(path, 'Prepared Data', 'weather_prediction_scaled_new.csv'))
```
